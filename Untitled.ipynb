```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('PSL-2016-2022.csv')
df.head()
```

<h1>BASIC STATISTIC</h1>

```python
df.describe()
```

<h1>Team performance analysis</h1>

```python
team_runs = df.groupby('team_1')['total_runs'].sum()
team_wickets = df.groupby('team_1')['wickets'].sum()
team_performance = pd.DataFrame({'Total Runs': team_runs, 'Total Wickets': team_wickets})
team_performance.plot(kind='bar')
plt.title('Team Performance Analysis')
plt.xlabel('Team')
plt.ylabel('Total Runs/Wickets')
plt.show()
```

<h1>Time series analyis</h1>

```python
innings_df = df.groupby(['inning', 'over'])['runs'].sum().unstack()
innings_df.plot(marker='o')
plt.title('Runs Scored in Each Over')
plt.xlabel('Over')
plt.ylabel('Total Runs')
plt.legend(title='Inning')
plt.show()
```

<h1>WIN LOSS ANALYSIOS</h1>

```python
team_wins = df[df['result'] == 'Winner'].groupby(['psl_year', 'team_1'])['result'].count().reset_index()
team_losses = df[df['result'] == 'Loser'].groupby(['psl_year', 'team_2'])['result'].count().reset_index()

team_wins.columns = ['psl_year', 'team', 'wins']
team_losses.columns = ['psl_year', 'team', 'losses']
team_performance = team_wins.merge(team_losses, on=['psl_year', 'team'], how='outer')
team_performance['wins'].fillna(0, inplace=True)
team_performance['losses'].fillna(0, inplace=True)
team_performance['win_percentage'] = (team_performance['wins'] / (team_performance['wins'] + team_performance['losses'])) * 100

plt.figure(figsize=(12, 6))
sns.lineplot(data=team_performance, x='psl_year', y='win_percentage', hue='team')
plt.title('Team Win-Loss Analysis Over the Years')
plt.xlabel('PSL Year')
plt.ylabel('Win Percentage')
plt.legend(title='Team')
plt.show()
```

```python
team_wins = df[df['result'] == 'Winner'].groupby(['psl_year', 'team_1'])['result'].count().reset_index()
team_losses = df[df['result'] == 'Loser'].groupby(['psl_year', 'team_2'])['result'].count().reset_index()

# Combine win and loss data to calculate overall performance
team_wins.columns = ['psl_year', 'team', 'wins']
team_losses.columns = ['psl_year', 'team', 'losses']
team_performance = team_wins.merge(team_losses, on=['psl_year', 'team'], how='outer')
team_performance['wins'].fillna(0, inplace=True)
team_performance['losses'].fillna(0, inplace=True)
team_performance['win_percentage'] = (team_performance['wins'] / (team_performance['wins'] + team_performance['losses'])) * 100

# Visualization: Team Win-Loss Analysis over the Years
plt.figure(figsize=(12, 6))
for team_name, team_data in team_performance.groupby('team'):
    plt.plot(team_data['psl_year'], team_data['win_percentage'], label=team_name)
plt.title('Team Win-Loss Analysis Over the Years')
plt.xlabel('PSL Year')
plt.ylabel('Win Percentage')
plt.legend(title='Team')
plt.show()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Match Result Analysis
result_counts = df['result'].value_counts()
result_counts.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Match Result Distribution')
plt.ylabel('')
plt.show()
```

<h1>Team batting analysis<h1>

```python
# Batting Analysis (Batting Average)
team_batting_avg = df.groupby(['psl_year', 'team_1'])['runs'].mean().reset_index()

# Visualization: Team Batting Average over the Years
plt.figure(figsize=(12, 6))
sns.lineplot(data=team_batting_avg, x='psl_year', y='runs', hue='team_1', label='Batting Average')
plt.title('Team Batting Average Over the Years')
plt.xlabel('PSL Year')
plt.ylabel('Batting Average')
plt.legend(title='Team')
plt.show()
```
